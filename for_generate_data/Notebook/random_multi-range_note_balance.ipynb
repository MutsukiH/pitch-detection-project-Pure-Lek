```python
# กำหนดโน๊ตในช่วง octave 4
octave_4_notes = [
    "C",
    "C#",
    "D",
    "D#",
    "E",
    "F",
    "F#",
    "G",
    "G#",
    "A",
    "A#",
    "B",
]


# train = pd.read_csv('../../data/ref_multi_note_600_tolerance_1(2).csv')


# ฟังก์ชันสุ่มลำดับโน๊ตโดยให้สมดุลและไม่ซ้ำ
def generate_balanced_sequences(note_count, sequence_count, tolerance):
    sequences = []
    note_frequency = [defaultdict(int) for _ in range(note_count)]
    samples_per_note = math.floor(sequence_count / len(octave_4_notes))

    # สร้างลำดับโน๊ตให้ครบตามจำนวนที่ต้องการ
    while len(sequences) < sequence_count:
        # sequence = [random.choice(octave_4_notes) for _ in range(note_count)]
        # print(sequence)
        # print(note_frequency)
        # ตรวจสอบความสมดุลของโน๊ตในลำดับและความไม่ซ้ำ
        # print(sequence)
        # คัดกรองโน๊ตที่จำนวนยังไม่ถึง samples_per_note
        available_notes = []
        for position in range(note_count):
            available_note = [
                note
                for note in octave_4_notes
                if note_frequency[position][note] < samples_per_note
            ]

            # หากไม่มีโน๊ตที่ยังไม่ถึงจำนวนที่ต้องการ ให้เพิ่มค่าความคลาดเคลื่อน
            if not available_note:
                available_note = [
                    note
                    for note in octave_4_notes
                    if note_frequency[position][note] < samples_per_note + tolerance
                ]

            available_notes.append(available_note)

        sequence = [
            random.choice(note_each_position) for note_each_position in available_notes
        ]
        # print(sequence)
        if (sequence not in sequences) and all(
            note_frequency[i][note] < samples_per_note
            for i, note in enumerate(sequence)
        ):
            sequences.append(sequence)
            for i, note in enumerate(sequence):
                note_frequency[i][note] += 1

            # print(
            #     available_notes,
            #     len(sequences),
            #     sequence,
            #     [str(note_frequency[i]) for i, note in enumerate(sequence)],
            # )

        elif (
            (sequence not in sequences)
            and len(sequences) >= samples_per_note * len(octave_4_notes)
            and all(
                note_frequency[i][note] < samples_per_note + tolerance
                for i, note in enumerate(sequence)
            )
        ):
            # print(len(sequences))
            sequences.append(sequence)
            for i, note in enumerate(sequence):
                note_frequency[i][note] += 1

        # print('-'*30)

    return list(sequences)


# สร้างและบันทึกผลลัพธ์
all_sequences = []
tolerance = 1
for note_count in range(3, 11):
    # print(f"{note_count} --------------------------------------------------")
    sequences = generate_balanced_sequences(note_count, 75, tolerance)
    all_sequences.extend(sequences)

# บันทึกผลลัพธ์ลงไฟล์
# with open("piano_sequences.txt", "w") as f:
#     for sequence in all_sequences:
#         f.write(" ".join(sequence) + "\n")

sampled_files_df = pd.DataFrame({"label": [x for x in all_sequences]})
sampled_files_df.to_csv("../../data/ref_list_multi_note_600_train.csv", index=False)
```

# สุ่มตัว validate

```python
data_train = pd.read_csv('../../data/ref_list_multi_note_600_train.csv')
data_train['label'] = data_train['label'].apply(ast.literal_eval)
data_train = data_train['label'].tolist()
```

```python
# กำหนดโน๊ตในช่วง octave 4
octave_4_notes = [
    "C",
    "C#",
    "D",
    "D#",
    "E",
    "F",
    "F#",
    "G",
    "G#",
    "A",
    "A#",
    "B",
]

# ฟังก์ชันสุ่มลำดับโน๊ตโดยให้สมดุลและไม่ซ้ำ
def generate_balanced_sequences_2(note_count, sequence_count, tolerance):
    sequences = []
    note_frequency = [defaultdict(int) for _ in range(note_count)]
    samples_per_note = math.floor(sequence_count / len(octave_4_notes))

    # สร้างลำดับโน๊ตให้ครบตามจำนวนที่ต้องการ
    while len(sequences) < sequence_count:

        available_notes = []
        for position in range(note_count):
            available_note = [
                note
                for note in octave_4_notes
                if note_frequency[position][note] < samples_per_note
            ]

            # หากไม่มีโน๊ตที่ยังไม่ถึงจำนวนที่ต้องการ ให้เพิ่มค่าความคลาดเคลื่อน
            if not available_note:
                available_note = [
                    note
                    for note in octave_4_notes
                    if note_frequency[position][note] < samples_per_note + tolerance
                ]

            available_notes.append(available_note)

        sequence = [
            random.choice(note_each_position) for note_each_position in available_notes
        ]
        # print(sequence)
        if (
            (sequence not in sequences)
            and (sequence not in data_train)
            and all(
                note_frequency[i][note] < samples_per_note
                for i, note in enumerate(sequence)
            )
        ):
            sequences.append(sequence)
            for i, note in enumerate(sequence):
                note_frequency[i][note] += 1

        elif (
            (sequence not in sequences)
            and len(sequences) >= samples_per_note * len(octave_4_notes)
            and (sequence not in data_train)
            and all(
                note_frequency[i][note] < samples_per_note + tolerance
                for i, note in enumerate(sequence)
            )
        ):
            # print(len(sequences))
            sequences.append(sequence)
            for i, note in enumerate(sequence):
                note_frequency[i][note] += 1

        # print('-'*30)

    return list(sequences)


# สร้างและบันทึกผลลัพธ์
all_sequences = []
tolerance = 1
for note_count in range(3, 11):
    # print(f"{note_count} --------------------------------------------------")
    sequences = generate_balanced_sequences_2(note_count, 25, tolerance)
    all_sequences.extend(sequences)

# บันทึกผลลัพธ์ลงไฟล์
# with open("piano_sequences.txt", "w") as f:
#     for sequence in all_sequences:
#         f.write(" ".join(sequence) + "\n")

sampled_files_df = pd.DataFrame({"label": [x for x in all_sequences]})
sampled_files_df.to_csv("../../data/ref_list_multi_note_200_validate.csv", index=False)
```
